```python
# Perso
import sys
from object_detector import *

import numpy as np
import pandas as pd

# Model path
model_path = "../models/craters_7.model"

device = 'cuda:0'

# Data
Xtest = torch.tensor(np.load('../data/data_test_mini.npy'))
Ytest = torch.tensor(pd.read_csv('../data/labels_test_mini.csv').values)
```

```python
object_detector = ObjectDetector()
object_detector.net.load_state_dict(torch.load(model_path))
```

```python
data = CraterDataset(Xtest, 8, Ytest)
```

```python
for X, idx in data.loader:
    break
```

```python
loc, conf, prior = object_detector.net(X.to(device))
to_proba = nn.Sigmoid()

for image_idx in range(X.shape[0]):

    current_label_idx = idx[image_idx]
    true_circles = data.Ytrain[data.Ytrain[:, 0] == current_label_idx, 1:4]

    proba = to_proba(conf)[image_idx] 
    predicted_circles = decode_location(loc[image_idx], prior) * 224

    best_predictions = predicted_circles[proba.sort(dim=0, descending=True)[1]].squeeze(1)[:10, :]

    transform = Grayscale(num_output_channels=1)
    to_image = ToPILImage()
    to_tensor = ToTensor()

    to_plot = to_tensor(transform(to_image(X[image_idx]))).squeeze(0)

    import matplotlib.pyplot as plt
    %matplotlib inline 

    fig, ax = plt.subplots(figsize=(10, 10))
    ax.imshow(to_plot, cmap=plt.cm.gray)

    # Draw the true boxes in green
    for y, x, r in true_circles:
        ax.add_patch(plt.Circle(
            (x, y), r, color='green', fill=False, linewidth=2))

    # Draw the predicted boxes in red
    for y, x, r in best_predictions:
        ax.add_patch(plt.Circle(
            (x, y), r, color='red', fill=False, linewidth=2))
        
    plt.show()

    print(proba.sort(dim=0, descending=True)[0][:10].view(-1))
```
