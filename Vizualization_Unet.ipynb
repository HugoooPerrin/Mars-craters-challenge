```python
# Perso
import sys
from object_detector_Unet import *

import numpy as np
import pandas as pd

# Model path
model_path = "../models/craters_9.model"

device = 'cuda:0'

# Data
Xtest = np.load('../data/data_test.npy')
Ytest = pd.read_csv('../data/labels_test.csv').values

# Xtest = np.load('../data/data_train.npy')
# Ytest = pd.read_csv('../data/labels_train.csv').values

# Keeping only pictures that includes a crater for training
with_crater = np.isin(np.arange(Xtest.shape[0]), np.unique(Ytest[:, 0]))
Xtest = Xtest[with_crater]

# Keeping y index coherent
replace = dict([(k, v) for k, v in zip(np.unique(Ytest[:, 0]), np.arange(Ytest.shape[0]))])
Ytest[:, 0] = [replace[i] for i in Ytest[:, 0]]
```

```python
object_detector = ObjectDetector()
object_detector.net.load_state_dict(torch.load(model_path))
```

```python
data = CraterDataset(Xtest, 8, Ytest)
```

```python
for X, idx in data.loader:
    break

X = X.to(device)
conf = object_detector.net(X)
to_proba = nn.Sigmoid()
```

```python
prior = PriorCircle(config).forward()
```

```python
for image_idx in range(X.shape[0]):

    current_label_idx = idx[image_idx]
    true_circles = data.Ytrain[data.Ytrain[:, 0] == current_label_idx, 1:4]

    proba = to_proba(conf)[image_idx] 

    best_priors = prior[proba.sort(dim=0, descending=True)[1]].squeeze(1)[:10, :] * 224

    to_plot = X[image_idx].squeeze(0)

    import matplotlib.pyplot as plt
    %matplotlib inline 

    fig, ax = plt.subplots(figsize=(10, 10))
    ax.imshow(to_plot, cmap=plt.cm.gray)

    # Draw the true boxes in green
    for y, x, r in true_circles:
        ax.add_patch(plt.Circle(
            (x, y), r, color='green', fill=False, linewidth=2))
        
    # Draw the best priors in blue
    for y, x, r in best_priors:
        ax.add_patch(plt.Circle(
            (x, y), r, color='blue', fill=False, linewidth=2))
        
    plt.show()

    print(proba.sort(dim=0, descending=True)[0][:10].view(-1))
```
